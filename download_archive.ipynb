```python
from datetime import date, timedelta
puzzle_dates = [
    (date.fromisoformat('2025-08-18') + timedelta(days=i)).strftime('%Y-%m-%d')
    for i in range((date.fromisoformat('2025-11-13') - date.fromisoformat('2025-08-18')).days + 1)
]
puzzle_dates
```

```python
all_puzzles = []
for puzzle_date in puzzle_dates:
    response = supabase.table("puzzles").select("print_date, type, editor, easy_puzzle, medium_puzzle, hard_puzzle").eq("print_date", puzzle_date).eq("type", 'pips').single().execute()

    all_puzzles.append(json.dumps(response.data, indent=2))
```

```python
all_puzzles = [json.loads(p) for p in all_puzzles]
```

```python
with open('all_puzzles.json','w',encoding='utf8') as f:
    json.dump(all_puzzles,f,indent=4)
```
